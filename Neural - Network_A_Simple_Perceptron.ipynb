Que 2. Explain the basic architecture and functioning of a Perceptron. What are its limitations?

Ans - Perceptron: Basic Architecture and Functioning

The Perceptron is one of the earliest artificial neural network models, introduced by Frank Rosenblatt in 1958. It is a single-layer binary classification model.

1️⃣ Basic Architecture of a Perceptron

A perceptron consists of:
1. Input Layer
- Inputs: x1, x2, x3, ...., xn
- Each input represents a feature.

2. Weights
- Each input has an associated weight: w1, w2, w3, ...., wn
- Weights determine the importance of each feature.

3. Bias (b)
- A constant value added to shift the decision boundary.

4. Summation Function
- Calculates weighted sum:

5. Activation Function (Step Function)
- Applies a threshold:

2️⃣ How a Perceptron Works (Step-by-Step)
1. Multiply each input by its weight.
2. Add all weighted inputs.
3. Add the bias.
4. Pass the result through a step activation function.
5. Output is either 0 or 1 (binary classification).

Example:

If predicting whether an email is spam:
- Inputs = word frequencies
- Output = Spam (1) or Not Spam (0)

4️⃣ Limitations of Perceptron

❌ 1. Can Only Solve Linearly Separable Problems

It can only classify data that can be separated by a straight line (2D) or hyperplane (higher dimensions).

Example:

- Works for AND, OR logic gates
- Fails for XOR problem

The inability to solve XOR was highlighted by Marvin Minsky and Seymour Papert in their book Perceptrons (1969).

❌ 2. Single-Layer Model

- No hidden layers
- Cannot learn complex patterns

❌ 3. Binary Output Only

- Produces only two classes (0 or 1)
- Not suitable for multi-class problems directly

❌ 4. Uses Hard Threshold (Non-differentiable)

- Step function is not differentiable
- Limits optimization flexibility compared to modern neural networks

5️⃣ Summary (Exam-Friendly Conclusion)

The perceptron is the simplest neural network model consisting of input nodes, weights, bias, and a step activation function. It performs binary classification using a linear decision boundary. However, its major limitation is that it cannot solve non-linearly separable problems like XOR, which led to the development of multi-layer neural networks.

Ans - Difference Between Loss Function and Cost Function in Neural Networks

Although the terms are often used interchangeably, there is a slight conceptual difference.

1️⃣ Loss Function

🔹 Definition:

A loss function measures the error for a single training example.

It tells us how far the predicted output is from the actual output for one data point.

🔹 Purpose:
- Measures individual prediction error.
- Guides weight updates during training.

🔹 Example:

If:
- Actual value (y) = 5
- Predicted value (ŷ) = 3

Squared Error Loss:

L=(y−y^​)2=(5−3)2=4

This value (4) is the loss for that single example.

2️⃣ Cost Function

🔹 Definition:

A cost function is the average (or total) of losses over the entire training dataset.

It gives the overall performance of the model.

🔹 Purpose:
- Measures total model error.
- Used to optimize model parameters.

Comparison of Gradient Descent, RMSprop, and Adam -

1️⃣ Gradient Descent

🔹 Concept:

Updates weights in the direction of the negative gradient.

🔹 Types:
- Batch Gradient Descent
- Stochastic Gradient Descent (SGD)
- Mini-batch Gradient Descent

✔ Advantages:
- Simple to understand and implement.
- Works well for convex problems.

❌ Disadvantages:
- Slow convergence.
- Sensitive to learning rate.
- Can get stuck in local minima.
- Same learning rate for all parameters.

2️⃣ RMSprop (Root Mean Square Propagation)

🔹 Concept:

Adjusts learning rate individually for each parameter using a moving average of squared gradients.

✔ Advantages:
- Faster convergence than basic GD.
- Handles non-stationary objectives well.
- Good for deep networks.

❌ Disadvantages:
- Requires tuning hyperparameters.
- Does not use momentum directly.

3️⃣ Adam (Adaptive Moment Estimation)

Adam combines:
- Momentum (like SGD with momentum)
- Adaptive learning rates (like RMSprop)

It uses:
- First moment (mean of gradients)
- Second moment (variance of gradients)

✔ Advantages:
- Fast convergence.
- Works well in most practical problems.
- Requires less tuning.
- Very popular in deep learning.

❌ Disadvantages:
- Can sometimes generalize worse than SGD.
- Slightly more memory usage.

Comparison Table -

| Feature       | Gradient Descent        | RMSprop             | Adam          |
| ------------- | ----------------------- | ------------------- | ------------- |
| Learning Rate | Same for all parameters | Adaptive            | Adaptive      |
| Uses Momentum | ❌ No (basic GD)         | ❌ No                | ✅ Yes         |
| Speed         | Slow                    | Faster              | Very Fast     |
| Memory Usage  | Low                     | Medium              | Higher        |
| Best For      | Simple problems         | RNNs, Deep networks | Most DL tasks |

Simple Understanding -
- Gradient Descent → Basic and simple
- RMSprop → Smarter learning rate adjustment
- Adam → Combination of Momentum + RMSprop (most powerful in practice)

Final Conclusion (Exam-Friendly) -

Optimizers play a crucial role in neural networks by minimizing the loss function through weight updates. Gradient Descent is simple but slow, RMSprop improves convergence using adaptive learning rates, and Adam combines momentum with adaptive learning for faster and more efficient training.

Que 6. Write a Python program to implement a single-layer perceptron from
scratch using NumPy to solve the logical AND gate.

- Use NumPy, Matplotlib, and Tensorflow/Keras for implementation.

Ans - Below is a complete Python program that:

- ✅ Implements a single-layer perceptron from scratch using NumPy
- ✅ Solves the AND gate
- ✅ Visualizes results using Matplotlib
- ✅ Implements the same using TensorFlow/Keras

1️⃣ Perceptron From Scratch (Using NumPy)


```python
import numpy as np
import matplotlib.pyplot as plt

# AND gate dataset
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 0, 0, 1])  # AND output

# Initialize weights and bias
np.random.seed(42)
weights = np.random.randn(2)
bias = np.random.randn()

learning_rate = 0.1
epochs = 20

# Step activation function
def step_function(z):
    return 1 if z >= 0 else 0

# Training
for epoch in range(epochs):
    for i in range(len(X)):
        z = np.dot(X[i], weights) + bias
        y_pred = step_function(z)
        error = y[i] - y_pred

        # Update weights and bias
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

print("Final Weights:", weights)
print("Final Bias:", bias)

# Testing
print("\nPredictions:")
for i in range(len(X)):
    z = np.dot(X[i], weights) + bias
    print(X[i], "->", step_function(z))
```

2️⃣ Decision Boundary Visualization (Matplotlib)

```python
# Plot data points
for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i][0], X[i][1], color='red')
    else:
        plt.scatter(X[i][0], X[i][1], color='blue')

# Decision boundary line
x_values = np.linspace(-0.5, 1.5, 100)
y_values = -(weights[0] * x_values + bias) / weights[1]

plt.plot(x_values, y_values)
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.xlabel("Input 1")
plt.ylabel("Input 2")
plt.title("Perceptron Decision Boundary for AND Gate")
plt.show()
```

3️⃣ Using TensorFlow / Keras

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# AND dataset
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
], dtype=float)

y = np.array([0, 0, 0, 1], dtype=float)

# Build single-layer perceptron model
model = keras.Sequential([
    layers.Dense(1, input_dim=2, activation='sigmoid')
])

model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train model
model.fit(X, y, epochs=100, verbose=0)

# Evaluate
loss, accuracy = model.evaluate(X, y, verbose=0)
print("\nTensorFlow Model Accuracy:", accuracy)

# Predictions
print("\nTensorFlow Predictions:")
predictions = model.predict(X)
for i in range(len(X)):
    print(X[i], "->", round(predictions[i][0]))
```

The perceptron correctly learns the AND gate, since AND is linearly separable.

Que 7. Implement and visualize Sigmoid, ReLU, and Tanh activation functions
using Matplotlib.

Ans - Here is a simple Python program to implement and visualize Sigmoid, ReLU, and Tanh activation functions using NumPy and Matplotlib.



```python
# Code :

import numpy as np
import matplotlib.pyplot as plt

# Create input values
x = np.linspace(-10, 10, 400)

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# -----------------------------
# Plot Sigmoid Function
# -----------------------------
plt.figure()
plt.plot(x, sigmoid(x))
plt.title("Sigmoid Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()

# -----------------------------
# Plot ReLU Function
# -----------------------------
plt.figure()
plt.plot(x, relu(x))
plt.title("ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()

# -----------------------------
# Plot Tanh Function
# -----------------------------
plt.figure()
plt.plot(x, tanh(x))
plt.title("Tanh Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

What This Code Does -

- Generates input values from -10 to 10
- Defines:
  - Sigmoid → Output between (0, 1)
  - ReLU → Output between [0, ∞)
  - Tanh → Output between (-1, 1)
- Plots each function separately

What You’ll Observe -
- Sigmoid → S-shaped curve
- ReLU → 0 for negative values, linear for positive
- Tanh → S-shaped curve centered at 0

Que 8. Use Keras to build and train a simple multilayer neural network on the
MNIST digits dataset. Print the training accuracy.

Ans - Here is a complete example using TensorFlow/Keras to build and train a simple Multilayer Neural Network (MLP) on the MNIST handwritten digits dataset, and print the training accuracy.

```python
# Code :

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np

# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()

# Normalize pixel values (0–255 → 0–1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# Flatten 28x28 images into 784-dimensional vectors
X_train = X_train.reshape(-1, 28 * 28)
X_test = X_test.reshape(-1, 28 * 28)

# Build Multilayer Perceptron (MLP) model
model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')  # 10 classes (0–9 digits)
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
history = model.fit(X_train, y_train, epochs=5, batch_size=32, verbose=1)

# Print training accuracy
train_loss, train_accuracy = model.evaluate(X_train, y_train, verbose=0)
print("\nTraining Accuracy:", train_accuracy)
```

Explanation -

1️⃣ Dataset

- MNIST contains 60,000 training images and 10,000 test images.
- Each image is 28×28 pixels.
- 10 classes: digits 0–9.

2️⃣ Model Architecture

- Input layer: 784 neurons (flattened image)
- Hidden Layer 1: 128 neurons (ReLU)
- Hidden Layer 2: 64 neurons (ReLU)
- Output Layer: 10 neurons (Softmax)

3️⃣ Optimizer

- Adam optimizer (adaptive and efficient)
